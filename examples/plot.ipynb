```python

from subgraph_counting import datasets, graph

PLOT_ROOT = "../data/figures"
```

```python
g = graph.Graph(
    datasets.get_zachary_graph(),
    True,
    node_attribute="x",
    edge_attribute="edge_attr",
)
node_font_size = 5
edge_font_size = 3

g.plot(
    f"{PLOT_ROOT}/_test_.svg",
    node_colors=False,
    node_labels=True,
    edge_labels=True,
    visual_style={
        "layout": "circle",
        "vertex_size": 0.1,
        "vertex_label_size": node_font_size,
        "edge_width": 0.5,
        "edge_label_size": edge_font_size,
    },
)
```

```python
from subgraph_counting import pattern, graph, conversion
for i in [3, 6]:
    g = graph.Graph(
        conversion.igraph_to_pyg_graph(pattern.create_cycle(i)),
        True
    )
    node_font_size = 5
    edge_font_size = 3

    g.plot(
        f"{PLOT_ROOT}/cycle-{i}.svg",
        node_colors=False,
        node_labels=True,
        edge_labels=True,
        visual_style={
            "layout": "circle",
            "vertex_size": 0.1,
            "vertex_label_size": node_font_size,
            "edge_width": 0.5,
            "edge_label_size": edge_font_size,
            "vertex_color": "white"
        },
    )
```
